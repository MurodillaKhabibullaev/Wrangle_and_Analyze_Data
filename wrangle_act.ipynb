## Cleaning step for archive data

## Define

- drop the rows with non-null values in 'in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id' and 'retweeted_status_timestamp' columns, because they are duplicated values. Then drop these columns too.

### Code

```python
# now we drop the rows indicated above
twit_archive_clean = twit_archive_clean.drop(twit_archive_clean[twit_archive_clean['in_reply_to_status_id'].notna()].index)
twit_archive_clean = twit_archive_clean.drop(twit_archive_clean[twit_archive_clean['retweeted_status_id'].notna()].index)

# let's quickly check for there isn't any
(twit_archive_clean['in_reply_to_status_id'].notna()|twit_archive_clean['retweeted_status_id']).sum()
```

```python
# ok, there is no left duplicates, now it's time to drop the columns
twit_archive_clean = twit_archive_clean.drop(columns = ['in_reply_to_status_id', 'in_reply_to_user_id', 'retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp'])
```

#### Test

```python
twit_archive_clean.info()
```

## Define

- change 'source' column so that containing just contents of HTML elements


### Code

```python
# now we will extract contents of html elements
from bs4 import BeautifulSoup

source_contents = []
for _ in range(twit_archive_clean['source'].shape[0]):
    soup = BeautifulSoup(twit_archive_clean['source'].iloc[_], 'html.parser')
    source_contents.append(soup.get_text())
twit_archive_clean['source'] = source_contents
```

#### Test

```python
twit_archive_clean.head(2)
```

## Define

- replace wrong names or 'None' values to 'Name not Given'

### Code

```python
# now we check values of name column for written in lower letters or 'None', and replace  them with 'Name not Given'
clean_names = []
for _ in range(twit_archive_clean['name'].shape[0]):
    if twit_archive_clean['name'].iloc[_].islower() or twit_archive_clean['name'].iloc[_] == 'None':
        twit_archive_clean['name'].iloc[_] = 'Name not Given'
    clean_names.append(twit_archive_clean['name'].iloc[_])
twit_archive_clean['name'] = clean_names
```

#### Test

```python
twit_archive_clean['name'].value_counts()
```

## Define

- add 'doggo', 'floofer', 'pupper' and 'puppo' columns into one 'dog_stages' column

### Code

```python
# now we add this column values to one column, first we need to change the values that is 'None' to ''

col_list = ['doggo', 'floofer', 'pupper', 'puppo']
for _ in col_list:
    twit_archive_clean.loc[twit_archive_clean[_] == 'None', _ ] = ''
    twit_archive_clean.loc[twit_archive_clean[_] == _, _ ] = ' ' + _ + ' '

# now it is time to adding this columns
twit_archive_clean['dog_stages'] = ''
for _ in col_list:
    twit_archive_clean['dog_stages']+= twit_archive_clean[_]

# if the dog_stages column has got more than one values, let's separate them by comma
twit_archive_clean['dog_stages'] = twit_archive_clean['dog_stages'].str.replace('  ', ', ')

# now it is turn to dropping unnecessary columns
twit_archive_clean = twit_archive_clean.drop(columns = col_list)
```

#### Test

```python
twit_archive_clean.sample(5)
```

```python
# what about building a linear model to predict if a comment get more favorites,
# what is the potential of its retweeting
import statsmodels.api as sm

df['intercept'] = 1

lm = sm.OLS(df['retweet_count'], df[['intercept', 'favorite_count']])
result = lm.fit()
result.summary()
```

>**According to this model** you can see that our model fit well. R-squared value close to one it is good. We can say every one unit increace in favorites, there is aproxomately 35 percent chance of being retweeted.
